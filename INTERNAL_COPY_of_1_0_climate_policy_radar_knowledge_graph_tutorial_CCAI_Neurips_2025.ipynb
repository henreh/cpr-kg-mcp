#Overview

<a name="overview"></a>

Climate Policy Radar helps people access and understand vast amounts of climate documents: laws, policies, NDCs, corporate transition plans, litigation documents, reports by statutory advisory bodies and industry bodies, and more. This tutorial is a dataset tutorial for:

- our open data: the full text and metadata of all of these documents, which we published open source.
- our 'concept store'. Climate documents are often long and filled with technical jargon. This makes them particularly difficult to analyse. Our concept store helps with this, giving users access to a rich web of expert-defined concepts and their relationships.

By linking this expert knowledge of climate change to the extensive curated database of climate documents, we show you how to create a climate policy knowledge graph. This can then be used in turn to analyse the global policy landscape.

After this tutorial you'll be able to download and understand our data (text and concepts), use the structure of our knowledge graph to train some simple but powerful classifiers, and do some introductory analysis of real climate policy documents.

<a name="climate-impact"></a>
# Climate Impact
Good decisions require access to timely, reliable data, models, and useful tools that make insight actionable. Yet there is a persistent gap in the climate sector of timely, credible, and connected information and intuitive tools to accelerate meaningful outcomes. The methods we show here help turn scattered, unstructured, long-text documents into accessible, usable and actionable shared intelligence – so that laws, policies, finance rules, corporate disclosures, litigation and more, can be understood and used by decision makers everywhere.

More specifically, we see two main roads towards impact: increased transparency and better, more evidence-based policy making. Transparency here plays a dual role: it allows outside actors to hold governments to account, while at the same time enhancing trust and therefore cooperation between governments [1, 2].

Both “policy learning” and “evidence-based policy making” are widely established and active research foci [e.g. 3, 4, 5]. The basic argument here is simple: to make better policies, you first need to understand the realities on the ground, including if and how any existing policies solving - or contributing to the problem. To see how this kind of thinking is also increasingly relevant in the climate discourse, we can look at the Intergovernmental Panel on Climate Change: the outline for their upcoming assessment report includes a whole chapter on “current adaptation progress, effectiveness and adequacy”, while on the mitigation side, overviews of existing policies and their effectiveness will be a mandated part for all the sectoral chapters [6]. Approaches such as what we describe here can form a great basis for such large-scale tracking and assessment work.

```python
def convert_to_python_native(value):
    if isinstance(value, np.ndarray):
        # we could use tolist() here but want the downstream values to be hashable
        return str(value)
    elif isinstance(value, np.integer):
        return int(value)
    elif isinstance(value, np.floating):
        return float(value)
    elif isinstance(value, np.bool_):
        return bool(value)
    return value


def get_document_count_by_column(column: str) -> pd.DataFrame:
    query = f"""
    SELECT "{column}", count(distinct document_id)
    FROM open_data
    GROUP BY "{column}"
    ORDER BY count(distinct document_id) DESC
    """

    result_df = db.sql(query).to_df()

    result_df = result_df.applymap(convert_to_python_native)

    result_df = result_df.rename(
        columns=lambda c: "count" if c.startswith("count") else c
    )

    return result_df

corpus_type_col = "document_metadata.corpus_type_name"

document_count_by_corpus_type = get_document_count_by_column(corpus_type_col)
document_count_by_corpus_type.set_index(corpus_type_col).plot(
    kind="pie", y="count", autopct="%1.1f%%", legend=False, ylabel=""
)
```

### Concepts and their relationships

We maintain the climate concepts we want to use in our knowledge graph in an instance of Wikibase (the software that powers Wikidata), at https://climatepolicyradar.wikibase.cloud/wiki/Main_Page.

This can be accessed via the Web UI, a set of APIs, or a graph query language called SPARQL. Here we'll have a look at some concepts and relationships between them using SPARQL, as we can do so without needing to authenticate.

First, we can have a look at the properties used in the knowledge graph and how often they're used. There are only a few – this has been intentional for now as we focus on adding more concepts. More specific relationship types will fall out later, as we learn how users use the graph.

```python
def sparql_query(query: str, timeout: int = 10) -> pd.DataFrame:
  """Run a SPARQL query against the concept store"""
  headers = {
      'User-Agent': 'Climate Policy Radar NeurIPS 2026 Tutorial',
      'Accept': 'application/sparql-results+json',
      'Content-Type': 'application/x-www-form-urlencoded'
  }
  endpoint_url = "https://climatepolicyradar.wikibase.cloud/query/sparql"

  response = requests.post(
      endpoint_url,
      data={'query': query},
      headers=headers,
      timeout=timeout
  )

  if response.status_code == 200:
      response_json = response.json()["results"]["bindings"]
      response_df = pd.DataFrame(response_json)
      response_df = response_df.map(lambda x: x["value"])
      return response_df

  else:
      raise Exception(f"Error response: {response.text}")

def get_all_properties():
  query = """
  SELECT ?prop ?propLabel (COUNT(?s) as ?count) WHERE {
    ?s ?p ?o .
    ?prop wikibase:directClaim ?p.
    FILTER(STRSTARTS(STR(?p), "https://climatepolicyradar.wikibase.cloud/prop/direct/"))
    SERVICE wikibase:label {
      bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en".
    }
  }
  GROUP BY ?prop ?propLabel
  ORDER BY DESC(?count)"""
  properties_raw = sparql_query(query)
  return properties_raw

properties_df = get_all_properties()
properties_df["prop"] = properties_df["prop"].apply(lambda x: re.findall(r"P\d+", x)[0])
properties_df
```

There are many more concepts. The next cell defines a miniature version of the `Concept` class in the [code we use to work with our knowledge graph](https://github.com/climatepolicyradar/knowledge-graph).

```python
class WikibaseID(str):
    """A Wikibase ID, which is a string that starts with a 'Q' followed by a number."""

    regex = r"^Q[1-9]\d*$"

    @classmethod
    def _validate(cls, value: str, field=None) -> str:
        """Validate that the Wikibase ID is in the correct format"""
        if not re.match(cls.regex, value):
            raise ValueError(f"{value} is not a valid Wikibase ID")
        return value

    @classmethod
    def __get_validators__(cls):
        """Return a generator of validators for the WikibaseID class"""
        yield cls._validate

    def __new__(cls, value: str) -> "WikibaseID":
        """Create a new instance of WikibaseID after validation"""
        validated_value = cls._validate(value)
        return str.__new__(cls, validated_value)

def url_to_wikibase_id(url: str) -> WikibaseID:
  candidates = re.findall(r"Q\d+", url)
  if len(candidates) == 1:
    return WikibaseID(candidates[0])

  raise ValueError("Invalid URL given")

class Concept(BaseModel):
  """Base class for a concept"""

  preferred_label: str = Field(
      ..., description="The preferred label for the concept", min_length=1
  )
  alternative_labels: list[str] = Field(
      default_factory=list, description="List of alternative labels for the concept"
  )
  negative_labels: list[str] = Field(
      default_factory=list,
      description=(
          "Labels which should not be matched instances of the concept. "
          "Negative labels should be unique, and cannot overlap with positive labels."
      ),
  )
  description: Optional[str] = Field(
      default=None,
      description=(
          "A short description of the concept which should be sufficient to "
          "disambiguate it from other concepts with similar labels"
      ),
  )
  definition: Optional[str] = Field(
      default=None,
      description=(
          "A more exhaustive definition of the concept, which should be enough for a "
          "human labeller to identify instances of the concept in a given text. "
      ),
  )
  wikibase_id: Optional[WikibaseID] = Field(
      default=None, description="The Wikibase ID for the concept"
  )
  subconcept_of: list[WikibaseID] = Field(
      default_factory=list,
      description="List of parent concept IDs",
  )
  has_subconcept: list[WikibaseID] = Field(
      default_factory=list, description="List of subconcept IDs"
  )
  related_concepts: list[WikibaseID] = Field(
      default_factory=list, description="List of related concept IDs"
  )

  @property
  def all_labels(self) -> list[str]:
      """Return a list of all unique labels for the concept"""
      return list(set([self.preferred_label] + self.alternative_labels))

  def get_concept_relationships(self):
    """
    Get all relationships for a single concept (both outgoing and incoming).

    Args:
        concept_id (str): The concept ID (e.g., "Q123" or full URI)
        max_relationships (int): Maximum number of relationships to retrieve

    Returns:
        pandas.DataFrame: DataFrame with columns for subject, predicate, object, and their labels
    """

    query = f"""
    PREFIX ent: <https://climatepolicyradar.wikibase.cloud/entity/>
    PREFIX dp: <https://climatepolicyradar.wikibase.cloud/prop/direct/>

    SELECT DISTINCT ?subject ?subjectLabel ?predicate ?predicateLabel ?object ?objectLabel ?direction WHERE {{
      {{
        # Outgoing relationships (concept as subject)
        BIND(ent:{self.wikibase_id} AS ?subject)
        ?subject ?p ?object .

        # Filter for direct properties only
        FILTER(STRSTARTS(STR(?p), "https://climatepolicyradar.wikibase.cloud/prop/direct/"))
      }}

      ?predicate wikibase:directClaim ?p.

      SERVICE wikibase:label {{
        bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en" .
      }}
    }}
    """

    relationships_df = sparql_query(query)
    return relationships_df

  @classmethod
  def load_from_wikibase(cls, concept_id: str) -> "Concept":
    """
    Load a concept from the concept store, which is hosted on Wikibase.

    Returns:
        Concept: concept populated with information from Wikibase
    """

    get_labels_aliases_sparql = f"""
    PREFIX ent: <https://climatepolicyradar.wikibase.cloud/entity/>
    PREFIX dp: <https://climatepolicyradar.wikibase.cloud/prop/direct/>
    SELECT ?item ?itemLabel ?itemAltLabel
      WHERE {{
        VALUES ?item {{ ent:{concept_id} }}
        SERVICE wikibase:label {{
          bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en".
          ?item rdfs:label ?itemLabel.
          ?item skos:altLabel ?itemAltLabel.
        }}
      }}"""

    results = sparql_query(get_labels_aliases_sparql)

    preferred_label = results["itemLabel"].values[0]
    alternative_labels = [a.strip() for a in results["itemAltLabel"].values[0].split(",")] if "itemAltLabel" in results.columns else []

    concept = Concept(
        preferred_label=preferred_label,
        alternative_labels=alternative_labels,
        description=None, # FIXME
        wikibase_id=concept_id
    )

    get_relationships_sparql = f"""
      PREFIX ent: <https://climatepolicyradar.wikibase.cloud/entity/>
      PREFIX dp: <https://climatepolicyradar.wikibase.cloud/prop/direct/>

      SELECT DISTINCT ?subject ?subjectLabel ?predicate ?predicateLabel ?object ?objectLabel ?direction WHERE {{
        {{
          # Outgoing relationships (concept as subject)
          BIND(ent:{concept_id} AS ?subject)
          ?subject ?p ?object .

          # Filter for direct properties only
          FILTER(STRSTARTS(STR(?p), "https://climatepolicyradar.wikibase.cloud/prop/direct/"))
        }}

        ?predicate wikibase:directClaim ?p.

        SERVICE wikibase:label {{
          bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en" .
        }}
      }}
      """

    relationships_df = sparql_query(get_relationships_sparql)

    objects_by_predicate = {
      name: group['object'].tolist()
      for name, group in relationships_df.groupby('predicateLabel', dropna=False)
    }

    concept.definition = objects_by_predicate.get("Definition", [None])[0]
    concept.related_concepts = [url_to_wikibase_id(u) for u in objects_by_predicate.get("Related to", [])]
    concept.has_subconcept = [url_to_wikibase_id(u) for u in objects_by_predicate.get("Has subconcept", [])]
    concept.subconcept_of = [url_to_wikibase_id(u) for u in objects_by_predicate.get("Subconcept of", [])]
    concept.negative_labels = objects_by_predicate.get("Negative labels")

    return concept

  def add_labels_from_recursive_subconcepts(self, depth: int = 3) -> "Concept":
      """Create a new concept with alternative labels from all subconcepts up to given depth"""

      if depth <= 0 or not self.has_subconcept:
          return self

      new_concept = copy.deepcopy(self)

      for concept_id in self.has_subconcept:
          subconcept = Concept.load_from_wikibase(concept_id)

          # Add labels from this subconcept
          subconcept_labels = [subconcept.preferred_label] + subconcept.alternative_labels
          new_concept.alternative_labels.extend(subconcept_labels)

          # Recursively get labels from subconcepts of this subconcept, up to depth
          if depth > 1:
              recursive_subconcept = subconcept.add_labels_from_recursive_subconcepts(depth - 1)
              new_concept.alternative_labels.extend(recursive_subconcept.alternative_labels)

      seen = set()
      new_concept.alternative_labels = [
          label for label in new_concept.alternative_labels
          if not (label in seen or seen.add(label))
      ]

      # Be nice to the endpoint
      time.sleep(0.1)

      return new_concept

  @property
  def wikibase_url(self) -> str:
    """Return the URL for the concept's Wikibase item"""
    return f"https://climatepolicyradar.wikibase.cloud/wiki/Item:{self.wikibase_id}"

  def to_markdown(self) -> str:
      """
      Return a complete representation of the concept in natural language

      ADAPTED AND SIMPLIFIED FROM https://github.com/climatepolicyradar/knowledge-graph.
      The original implementation contains rendering of the concept neighbourhood
      and labelled passages.

      Rather than outputting a machine-readable form of the concept (eg in JSON), the
      output should be human-readable, suitable for human or LLM consumption.

      :param WikibaseSession wikibase: A Wikibase session
      """
      sections = []

      # Title
      sections.append(f"# {self.preferred_label}")

      # Description and Definition
      if self.description:
          sections.append("## Description")
          sections.append(self.description)

      if self.definition:
          sections.append("## Definition")
          sections.append(self.definition)

      # Labels
      if self.alternative_labels:
          sections.append(
              "## Alternative labels, synonyms, acronyms, and related terms"
          )
          sections.append(
              "\n".join(f"- {label}" for label in self.alternative_labels)
          )

      if self.negative_labels:
          sections.append("## Not to be confused with")
          sections.append("\n".join(f"- {label}" for label in self.negative_labels))

      return "\n\n".join(sections)
```

We can visualise the network around a concept `n` hops away.

```python
ROOT_CONCEPT = Concept.load_from_wikibase("Q435") # Coral bleaching
N_HOPS = 2

def visualize_concept_network(root_concept, max_hops=2):
    """
    Create a force-directed graph of concepts within a certain number of hops.

    Args:
        root_concept (Concept): The root concept to start from
        max_hops (int): Maximum number of hops to traverse

    Returns:
        plotly Figure: Interactive network graph
    """

    nodes, edges, visited = {}, [], set()
    colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#95a5a6']

    def traverse(concept_id, hop):
        if hop > max_hops:
            return

        if concept_id in nodes and nodes[concept_id]['hop'] <= hop:
            return

        if concept_id not in visited:
            visited.add(concept_id)

            concept = root_concept if concept_id == root_concept.wikibase_id else Concept.load_from_wikibase(concept_id)

            nodes[concept_id] = {
                'label': concept.preferred_label,
                'hop': hop,
                'desc': concept.description or 'No description'
            }

            for sub in concept.has_subconcept:
                edges.append({'source': concept_id, 'target': sub})
                if hop < max_hops:
                    traverse(sub, hop + 1)

            for parent in concept.subconcept_of:
                edges.append({'source': parent, 'target': concept_id})
                if hop < max_hops:
                    traverse(parent, hop + 1)

            for rel in concept.related_concepts:
                edges.append({'source': concept_id, 'target': rel})
                if hop < max_hops:
                    traverse(rel, hop + 1)
        else:
            # Update hop distance if we found a shorter path
            if concept_id in nodes and hop < nodes[concept_id]['hop']:
                nodes[concept_id]['hop'] = hop

    traverse(root_concept.wikibase_id, 0)

    # Create networkx graph and prepare plot data
    G = nx.Graph()
    G.add_nodes_from(nodes.keys())
    valid_edges = [(e['source'], e['target']) for e in edges
                   if e['source'] in nodes and e['target'] in nodes]
    G.add_edges_from(valid_edges)
    pos = nx.spring_layout(G, k=2, seed=42)

    node_data = [(pos[nid][0], pos[nid][1], data['label'], data['hop'], data['desc'])
                 for nid, data in nodes.items()]
    node_x, node_y, labels, hops, descs = zip(*node_data) if node_data else ([], [], [], [], [])

    edge_x, edge_y = [], []
    for edge in edges:
        if edge['source'] in nodes and edge['target'] in nodes and edge['source'] in pos and edge['target'] in pos:
            x0, y0 = pos[edge['source']]
            x1, y1 = pos[edge['target']]
            edge_x.extend([x0, x1, None])
            edge_y.extend([y0, y1, None])

    # Create plot
    fig = go.Figure([
        go.Scatter(x=edge_x, y=edge_y, mode='lines', line=dict(width=1, color='#7f8c8d'),
                   hoverinfo='none', showlegend=False),
        go.Scatter(x=node_x, y=node_y, mode='markers+text', text=labels,
                   textfont=dict(size=10, color='#2c3e50'), textposition="middle center",
                   marker=dict(size=[25 if h==0 else 15 for h in hops],
                              color=[colors[min(h, 4)] for h in hops],
                              line=dict(width=2, color='#2c3e50')),
                   hovertemplate='<b>%{text}</b><br>Hops: %{customdata}<extra></extra>',
                   customdata=hops, showlegend=False,
                   marker_symbol='circle')
    ])

    fig.update_layout(
        title=f"Concept Network: {root_concept.preferred_label}",
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        plot_bgcolor='white', margin=dict(t=50, b=20, l=20, r=20),
        width=800, height=600,
        dragmode='pan',
        hovermode='closest'
    )

    legend_items = [("Root", colors[0]), ("1 Hop", colors[1]), ("2 Hops", colors[2]),
                   ("3 Hops", colors[3]), ("4+ Hops", colors[4])]

    for i, (label, color) in enumerate(legend_items[:max_hops + 2]):
        fig.add_annotation(
            x=0.02, y=0.98 - i*0.05,
            xref='paper', yref='paper',
            text=f'<span style="color:{color}">●</span> {label}',
            showarrow=False,
            font=dict(size=12),
            align="left"
        )

    return fig


visualize_concept_network(ROOT_CONCEPT, N_HOPS)
```

# Methodology

To create a knowledge graph on climate-relevant documents using the expert-defined concepts from the concept stores, we need to **create classifiers**. We can exploit the richness of the information in the concept store to create some surprisingly simple but effective classifiers, which scale to run on millions of text passages in little time and with a small resource impact.

We also introduce using the information contained in the concept store to create zero- and few-shot classifiers using large language models. LLMs applied to this dataset are powerful but can come with large resource implications and potential negative climate impact – we provide opportunity for discussing this balance in an open-ended task at the end of this tutorial.

## Creating a simple zero-shot classifier

First we create a `Span` model and base `Classifier` class, then a `KeywordClassifier` which uses the *preferred labels* and *alternative labels* in the concept store to build a simple classifier.

```python
class Span(BaseModel):
    """Represents a span within a text."""

    text: str = Field(..., description="The text of the span")
    start_index: int = Field(
        ..., ge=0, description="The start index of the span within the text"
    )
    end_index: int = Field(
        ..., gt=0, description="The end index of the span within the text"
    )
    concept_id: Optional[WikibaseID] = Field(
        None,
        description="The wikibase identifier associated with the span",
        examples=["Q42"],
    )


class Classifier(ABC):
  """Abstract class for all classifier types"""

  def __init__(
      self,
      concept: Concept,
      **kwargs
  ):
    self.concept = concept

  def fit(self, **kwargs) -> "Classifier":
    """
    Train the classifier on the data in the concept.

    This is a template method, which subclasses can override to implement their
    specific training logic. The base class is a no-op, allowing subclasses to
    optionally implement it, only when training is needed. Keeping the template
    method here provides a consistent interface across all classifiers while
    allowing flexibility in implementation.

    Args:
        **kwargs: Training configuration parameters. The specific parameters
            required will depend on the classifier implementation.

    Returns:
        Classifier: The trained classifier
    """

    return self

  @abstractmethod
  def predict(self, text: str) -> list[Span]:
      """
      Predict whether the supplied text contains an instance of the concept.

      :param str text: The text to predict on
      :return list[Span]: A list of spans in the text
      """
      raise NotImplementedError

  def predict_batch(self, texts: list[str]) -> list[list[Span]]:
    """
    Predict whether the supplied texts contain instances of the concept.

    :param list[str] texts: The texts to predict on
    :return list[list[Span]]: A list of spans in the texts for each text
    """
    return [self.predict(text) for text in texts]


class KeywordClassifier(Classifier):
  """
  Classifier uses keyword matching to find instances of a concept in text.

  ADAPTED FROM https://github.com/climatepolicyradar/knowledge-graph

  Keywords are based on the preferred and alternative labels of the concept. The
  classifier uses regular expressions to match the keywords in the text. Regexes are
  applied differently based on the casing of the original label:

  1. Case-sensitive matching: Applied to incoming labels containing any uppercase
  characters.
  2. Case-insensitive matching: Applied to labels containing only lowercase
  characters (this should apply to most labels).

  Regexes are applied at word boundaries in order of decreasing length to ensure that
  longer keywords are matched first.

  This approach allows for nuanced matching where:
  - Uppercase-containing labels (e.g., "WHO") will only match exactly ("WHO", not
  "who")
  - Lowercase-only labels (e.g., "virus") will match regardless of case ("virus",
  "Virus", "VIRUS")

  This distinction is particularly useful for differentiating between common words
  and specific entities (e.g., "who" vs "WHO" for the World Health Organization).
  """

  def __init__(self, concept: Concept):
        """
        Create a new KeywordClassifier instance.

        :param Concept concept: The concept which the classifier will identify in text
        """
        super().__init__(concept)

        def create_pattern(labels):
            return r"\b(?:" + "|".join(labels) + r")\b"

        self.case_sensitive_labels = []
        self.case_insensitive_labels = []

        # Sort labels by length in descending order so that longer labels are matched first
        sorted_labels = sorted(self.concept.all_labels, key=len, reverse=True)

        for label in sorted_labels:
            if label.strip():  # Ensure the label is not just whitespace
                if any(char.isupper() for char in label) or any(
                    ord(char) > 127 for char in label
                ):
                    # Labels including uppercase or non-ASCII characters are added to the case-sensitive list
                    self.case_sensitive_labels.append(re.escape(label))
                else:
                    # Only pure ASCII lowercase labels are added to the case-insensitive list
                    self.case_insensitive_labels.append(re.escape(label))

        # Case-sensitive pattern: matches exactly as provided
        self.case_sensitive_pattern = re.compile(
            create_pattern(self.case_sensitive_labels)
        )

        # Case-insensitive pattern: matches regardless of case
        self.case_insensitive_pattern = re.compile(
            create_pattern(self.case_insensitive_labels), re.IGNORECASE
        )

  def predict(self, text: str) -> list[Span]:
        """
        Predict whether the supplied text contains an instance of the concept.

        This method applies both case-sensitive and case-insensitive patterns to find
        matches. It ensures that:

        1. Longer matches take precedence over shorter ones
        2. No overlapping matches are returned
        3. Case-sensitive matches are found exactly as provided
        4. Case-insensitive matches can be found regardless of casing

        :param str text: The text to predict on
        :return list[Span]: A list of spans in the text
        """
        spans = []
        matched_positions = set()

        # Case-sensitive matching
        for match in self.case_sensitive_pattern.finditer(text):
            start, end = match.span()
            if start != end and not any(start <= p < end for p in matched_positions):
                spans.append(
                    Span(
                        text=text,
                        concept_id=self.concept.wikibase_id,
                        start_index=start,
                        end_index=end,
                    )
                )
                matched_positions.update(range(start, end))

        # Case-insensitive matching
        for match in self.case_insensitive_pattern.finditer(text):
            start, end = match.span()
            if start != end and not any(start <= p < end for p in matched_positions):
                spans.append(
                    Span(
                        text=text,
                        concept_id=self.concept.wikibase_id,
                        start_index=start,
                        end_index=end,
                    )
                )
                matched_positions.update(range(start, end))

        return spans

```

We can run the `KeywordClassifier` a document – let's train one for 'coral bleaching', and run it on the [Seychelles Wetlands Policy and Action Plan 2019-2022](https://app.climatepolicyradar.org/documents/seychelles-wetlands-policy-and-action-plan-2019-2022_d022).

```python
document_slug = "seychelles-wetlands-policy-and-action-plan-2019-2022_d022"
concept_id = "Q435" # Coral bleaching

document_text = get_text_from_document(document_slug)
print(f"Found {len(document_text)} passages")

print("Loading concept and classifier:")

concept = Concept.load_from_wikibase(concept_id)
classifier = KeywordClassifier(concept)

spans = classifier.predict_batch(document_text)

print("\nSpans:")
_ = [display(span) for span_group in spans for span in span_group]
```

In practice, we bolster classifiers by using the concept hierarchies to get more keywords for free. Let's try this with coral bleaching's parent concept, *marine risk*.

```python
concept_id = "Q368" # Marine risk

concept = Concept.load_from_wikibase(concept_id)
n_labels = len(concept.all_labels)
classifier_Q368 = KeywordClassifier(concept)
print(f"Concept {concept.preferred_label} has {n_labels} total labels")

concept_with_subconcept_labels = concept.add_labels_from_recursive_subconcepts(depth=3)
n_labels_after = len(concept.all_labels)
classifier_Q368_with_subconcept_labels = KeywordClassifier(concept_with_subconcept_labels)
print(f"Using the hierarchy, it now has {n_labels_after} total labels")
```

Running the two classifiers on the document above, we can see that the classifier with subconcept labels lets us obtain a much higher recall than without.



```python
spans_no_subconcepts = classifier_Q368.predict_batch(document_text)
spans_with_subconcepts = classifier_Q368_with_subconcept_labels.predict_batch(document_text)

n_spans_no_subconcepts = len([span for span_group in spans_no_subconcepts for span in span_group])
n_spans_with_subconcepts = len([span for span_group in spans_with_subconcepts for span in span_group])

print(f"{n_spans_no_subconcepts} without subconcepts; {n_spans_with_subconcepts} with\n\n")

[span.text for span_group in spans_with_subconcepts for span in span_group]
```

Hopefully by now you're convinced that the concept store contains a treasure trove of reliable, referenced information on climate concepts. As we can easily access this in code via the `Concept` class, we can provide this to an LLM to easily produce a zero-shot classifier.

In practice, we use a few sampled human labels to create a *few-shot* classifier.

We also don't use LLMs for classification in production due to the resource impact of running them on millions of text passages. Instead, we distil a `ModernBERT` model using a student-teacher framework. This is ideal in a production setting as we can debug these models using their training data, and relate each predicted `Span` to a frozen model instance.
